```python

import datetime

class Patient:
    def __init__(self, patient_id, name, age):
        self.patient_id = patient_id
        self.name = name
        self.age = age

class Doctor:
    def __init__(self, doctor_id, name, specialization):
        self.doctor_id = doctor_id
        self.name = name
        self.specialization = specialization

class Appointment:
    def __init__(self, patient, doctor, date_time):
        self.patient = patient
        self.doctor = doctor
        self.date_time = date_time

class Hospital:
    def __init__(self):
        self.patients = {}
        self.doctors = {}
        self.appointments = []

    def add_patient(self, patient):
        self.patients[patient.patient_id] = patient

    def add_doctor(self, doctor):
        self.doctors[doctor.doctor_id] = doctor

    def book_appointment(self, patient_id, doctor_id, date_time):
        if patient_id in self.patients and doctor_id in self.doctors:
            appointment = Appointment(self.patients[patient_id], self.doctors[doctor_id], date_time)
            self.appointments.append(appointment)
            print("Appointment booked successfully.")
        else:
            print("Error: Patient or Doctor not found.")

    def display_all(self):
        print("\n--- Patients ---")
        for p in self.patients.values():
            print(f"ID: {p.patient_id}, Name: {p.name}, Age: {p.age}")
        
        print("\n--- Doctors ---")
        for d in self.doctors.values():
            print(f"ID: {d.doctor_id}, Name: Dr. {d.name}, Spec.: {d.specialization}")

        print("\n--- Appointments ---")
        for a in self.appointments:
            print(f"Patient: {a.patient.name} with Dr. {a.doctor.name} on {a.date_time.strftime('%Y-%m-%d %H:%M')}")

if __name__ == "__main__":
    hospital = Hospital()

    
    hospital.add_patient(Patient("P001", "prerana", 35))
    hospital.add_patient(Patient("P002", "athrava", 50))
    hospital.add_doctor(Doctor("D001", "rakesh", "Cardiology"))


    appointment_time = datetime.datetime(2025, 8, 20, 10, 30)
    hospital.book_appointment("P001", "D001", appointment_time)

    
    hospital.display_all()

    

```
